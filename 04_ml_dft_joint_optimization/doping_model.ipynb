```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
make_doped_HfO2_supercells_unique.py
"""

import itertools
import random
from pathlib import Path

from pymatgen.core import Structure, Element
from pymatgen.analysis.structure_matcher import StructureMatcher

# -------------------------- 用户参数 --------------------------
cif_file = "HfO2.cif"
supercell_matrix = [[2, 0, 0],
                    [0, 2, 0],
                    [0, 0, 2]]

k = 3                 # 用 Y 替换的 Hf 个数
max_models = None     # None=全部；设整数则随机抽样
out_dir = Path(f"{k}Y_unique")
out_dir.mkdir(exist_ok=True)

# -------------------------- 读入并扩胞 ------------------------
orig = Structure.from_file(cif_file)
supercell = orig.copy()
supercell.make_supercell(supercell_matrix)

hf_indices = [i for i, s in enumerate(supercell) if s.species_string == "Hf"]
assert len(hf_indices) == 32, "超胞应含 32 个 Hf"

# ------------------------ 生成组合 ----------------------------
all_combos = list(itertools.combinations(hf_indices, k))
if max_models and max_models < len(all_combos):
    random.shuffle(all_combos)
    all_combos = all_combos[:max_models]

print(f"[INFO] 组合待检 {len(all_combos)} 条")

# ------------------ StructureMatcher 去重 ---------------------
matcher = StructureMatcher(
    ltol=0.2, stol=0.3, angle_tol=5,
    primitive_cell=False, scale=True,
    attempt_supercell=False, allow_subset=False
)

unique_structs = []
file_counter = 0

for combo in all_combos:
    model = supercell.copy()
    for idx in combo:
        model[idx] = Element("Y")

    if any(matcher.fit(u, model) for u in unique_structs):
        continue        # 对称等价，跳过

    unique_structs.append(model)
    file_counter += 1
    filename = out_dir / f"{k}Y_unique_{file_counter:03d}.cif"
    model.to(fmt="cif", filename=str(filename))
    print(f"  [SAVE] {filename.name}  (sites={combo})")

print(f"\n[DONE] 获得 {file_counter} 个对称不等价的 {k}Y 掺杂模型")
```
