```python
import csv
import re
from ase.db import connect

def extract_element_counts(formula):
    """提取化学式中的所有元素及其数量"""
    parsed_elements = re.findall(r'([A-Z][a-z]*)(\d*)', formula)
    element_counts = {}
    for elem, count in parsed_elements:
        element_counts[elem] = int(count) if count else 1
    return element_counts

# 连接数据库
db1 = connect('Surfaces_optimized.db')
db2 = connect('Surfaces_optimized_Li_opitmized.db')

# Open the CSV file for writing
with open('Li_opitmized.csv', 'w', newline='') as csvfile:
    fieldnames = ['Record', 'Formula_db1', 'Formula_db2', 'Adsorption_Energy']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    # Write the header
    writer.writeheader()

    # 遍历db1的所有记录
    for i, row_db1 in enumerate(db1.select(), 1):
        formula_db1 = row_db1.get('formula')
        counts_db1 = extract_element_counts(formula_db1)
        
        matched_row_db2 = None
        for row in db2.select():
            formula_db2 = row.get('formula')
            counts_db2 = extract_element_counts(formula_db2)
            
            # Subtract the appropriate Li and S counts
            counts_db2["Li"] = counts_db2.get("Li", 0) - 2
            counts_db2["S"] = counts_db2.get("S", 0) - 1
            
            if counts_db2["Li"] <= 0: del counts_db2["Li"]
            if counts_db2["S"] <= 0: del counts_db2["S"]

            if counts_db1 == counts_db2:
                matched_row_db2 = row
                break

        if matched_row_db2:
            energy_db1 = row_db1.get('energy')
            energy_db2 = matched_row_db2.get('energy')
            adsorption_energy = energy_db2 - energy_db1 + 0.29710367
        else:
            formula_db2 = None
            adsorption_energy = None
        
        # Write the data to the CSV file
        writer.writerow({
            'Record': i, 
            'Formula_db1': formula_db1, 
            'Formula_db2': formula_db2,
            'Adsorption_Energy': adsorption_energy
        })

print("The data has been written to 'Li_opitmized.csv'")
```
