```python
import h5py
import numpy as np

# 读取HDF5文件
with h5py.File('kappa-m111111.hdf5', 'r') as f:
    temperature = f['temperature'][:]

# 打印温度列表和索引
print("="*50)
print("温度索引对照表")
print("="*50)
print(f"{'索引':<8} {'温度 (K)':<12}")
print("-"*50)
for idx, temp in enumerate(temperature):
    marker = " ← 默认" if idx == 6 else ""
    print(f"{idx:<8} {temp:<12.1f}{marker}")
print("="*50)
print(f"\n总共 {len(temperature)} 个温度点")
print(f"温度范围: {temperature.min():.0f} - {temperature.max():.0f} K")
```
