```python
import os
import pandas as pd

# txt 파일 경로
txt_file_path = "./SJD4_18031-AI.txt"

# csv 파일 경로
csv_file_path = './Andor/Excitation.csv'


# 텍스트 파일이 있는 폴더 경로
folder_path = './Andor'
csv_folder_path = os.path.join(folder_path, 'csv_files')
os.makedirs(csv_folder_path, exist_ok=True)

# 폴더 내 파일명이 '1D'로 끝나는 모든 텍스트 파일을 찾기
txt_files = [f for f in os.listdir(folder_path) if f.endswith('1D.txt')]

# 파일 이름을 알파벳순으로 정렬
txt_files.sort()

# 각 텍스트 파일을 CSV 파일로 변환
for file in txt_files:
    file_path = os.path.join(folder_path, file)
    with open(file_path, 'r') as f:
        lines = f.readlines()
        # "X\tCounts" 라인의 인덱스를 찾기
        start_index = None
        for i, line in enumerate(lines):
            if line.startswith('X\tCounts'):
                start_index = i
                break
        
        if start_index is None:
            print(f"Warning: 'X\tCounts' header not found in {file}")
            continue
        
        # 해당 라인부터 끝까지 DataFrame으로 변환
        data = pd.read_csv(file_path, delimiter='\t', skiprows=start_index)
        
        # CSV 파일로 저장
        csv_file_path = os.path.join(csv_folder_path, file.replace('.txt', '.csv'))
        data.to_csv(csv_file_path, index=False, header=False)
        print(f'Converted {file} to CSV format.')

# 폴더 내 모든 변환된 CSV 파일을 찾기
csv_files = [os.path.join(csv_folder_path, f) for f in os.listdir(csv_folder_path) if f.endswith('.csv')]

# 파일 이름을 알파벳순으로 정렬
csv_files.sort()

# 첫 번째 CSV 파일을 기준으로 병합할 준비
base_file = csv_files[0]
base_df = pd.read_csv(base_file, header=None)
base_df.columns = ['X', os.path.basename(base_file).replace('.csv', '')]

# 나머지 CSV 파일들을 읽어서 두 번째 열을 병합
for file in csv_files[1:]:
    df = pd.read_csv(file, header=None)
    if df.shape[1] >= 2:  # 두 번째 열이 존재하는지 확인
        file_label = os.path.basename(file).replace('.csv', '')
        base_df[file_label] = df.iloc[:, 1]
    else:
        print(f"Warning: Second column not found in {file}")

# 첫 번째 행에 파일명을 추가
file_labels = ['X'] + [os.path.basename(file).replace('.csv', '') for file in csv_files]
header_df = pd.DataFrame([file_labels], columns=base_df.columns)

# 병합된 DataFrame을 새로운 CSV 파일로 저장
merged_df = pd.concat([header_df, base_df], ignore_index=True)
output_file_path = os.path.join(folder_path, 'merged_file.csv')
merged_df.to_csv(output_file_path, index=False, header=False)

print(f'Merged CSV file saved to {output_file_path}')

# 1번 csv 파일 경로
csv_file_1_path = './Andor/Excitation.csv'

# 2번 csv 파일 경로
csv_file_2_path = './Andor/merged_file.csv'

# 1번 csv 파일을 데이터프레임으로 읽어오기
df1 = pd.read_csv(csv_file_1_path, header=None)

# 1번 csv 파일의 3열 데이터 (1행~16행) 추출
column_data = df1.iloc[:16, 2].values

# 2번 csv 파일을 데이터프레임으로 읽어오기
df2 = pd.read_csv(csv_file_2_path, header=None)

# 2번 csv 파일의 1행 데이터 (2열~17열) 업데이트
df2.iloc[0, 1:17] = column_data

# 수정된 2번 csv 파일 저장하기
df2.to_csv(csv_file_2_path, header=False, index=False)

print(f"파일이 성공적으로 업데이트되어 {csv_file_2_path}로 저장되었습니다.")

# CSV 파일 읽기
input_file = './Andor/merged_file.csv'  # 입력 파일 이름
output_file = './Andor/merged_file.csv'  # 출력 파일 이름

# CSV 파일을 DataFrame으로 읽기
df = pd.read_csv(input_file)

# 17열(열 인덱스 16) 삭제
if df.shape[1] > 16:
    df.drop(df.columns[16], axis=1, inplace=True)
else:
    print("Warning: 17th column does not exist in the CSV file.")

# 첫 번째 열의 이름 가져오기
first_column = df.columns[0]

# 첫 번째 열 데이터 변환
df[first_column] = (df[first_column]*0.12666 + 449.99336) 

# 변환된 데이터 저장
df.to_csv(output_file, index=False)

print(f"변환된 데이터를 {output_file}에 저장했습니다.")
```

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from scipy.interpolate import griddata
import os

def get_excitation_energy_mapping(excitation_csv_path="./Andor/Excitation.csv"):
    """从Excitation.csv文件获取文件名到激发能量的映射关系"""
    try:
        # 读取Excitation.csv文件
        df = pd.read_csv(excitation_csv_path)
        
        # 检查必要的列是否存在
        if 'Filename' not in df.columns or 'BL 8 Energy' not in df.columns:
            print("Error: 'Filename' or 'BL 8 Energy' column not found in Excitation.csv")
            return {}
        
        # 创建映射字典
        energy_mapping = {}
        for index, row in df.iterrows():
            filename = row['Filename']
            energy = row['BL 8 Energy']
            
            # 处理文件名，添加-1D后缀（如果需要）
            if not filename.endswith('-1D'):
                filename_key = filename + '-1D'
            else:
                filename_key = filename
            
            energy_mapping[filename_key] = energy
        
        print(f"成功从 {excitation_csv_path} 读取了 {len(energy_mapping)} 个激发能量映射")
        print("映射关系:")
        for key, value in energy_mapping.items():
            print(f"  {key}: {value:.8f} eV")
        
        return energy_mapping
        
    except FileNotFoundError:
        print(f"错误: 找不到文件 {excitation_csv_path}")
        print("使用默认的硬编码映射关系...")
        return get_default_energy_mapping()
    except Exception as e:
        print(f"读取Excitation.csv时出错: {str(e)}")
        print("使用默认的硬编码映射关系...")
        return get_default_energy_mapping()

def get_default_energy_mapping():
    """默认的激发能量映射关系（作为备选）"""
    energy_mapping = {
        'SJD4_18031-00001-1D': 529.90157898,
        'SJD4_18031-00002-1D': 530.10417319,
        'SJD4_18031-00003-1D': 530.31199310,
        'SJD4_18031-00004-1D': 530.50475223,
        'SJD4_18031-00005-1D': 530.70780803,
        'SJD4_18031-00006-1D': 530.92626647,
        'SJD4_18031-00007-1D': 531.10930027,
        'SJD4_18031-00008-1D': 531.30263958,
        'SJD4_18031-00009-1D': 531.50630679,
        'SJD4_18031-00010-1D': 531.71013021,
        'SJD4_18031-00011-1D': 531.90900862,
        'SJD4_18031-00012-1D': 532.12845734,
        'SJD4_18031-00013-1D': 532.30210317,
        'SJD4_18031-00014-1D': 532.50653755,
        'SJD4_18031-00015-1D': 532.71112902,
        'SJD4_18031-00016-1D': 532.91587776
    }
    return energy_mapping

def match_columns_with_energies(data_columns, energy_mapping):
    """
    匹配merged_file.csv中的列名与激发能量映射
    处理列名可能是文件名或能量值的情况
    """
    matched_pairs = []
    
    # 创建能量值到文件名的反向映射
    energy_to_filename = {v: k for k, v in energy_mapping.items()}
    
    print(f"\n=== 列名匹配过程 ===")
    print(f"merged_file.csv中的列名: {data_columns}")
    
    for col in data_columns:
        matched = False
        
        # 方法1: 直接匹配文件名
        if col in energy_mapping:
            matched_pairs.append((col, energy_mapping[col]))
            print(f"✓ 直接匹配: {col} -> {energy_mapping[col]:.8f} eV")
            matched = True
        
        # 方法2: 尝试作为能量值匹配
        if not matched:
            try:
                energy_value = float(col)
                # 在能量映射中查找最接近的能量值
                best_match = None
                min_diff = float('inf')
                
                for filename, energy in energy_mapping.items():
                    diff = abs(energy - energy_value)
                    if diff < min_diff:
                        min_diff = diff
                        best_match = (filename, energy)
                
                # 如果差值小于0.001 eV，认为是匹配的
                if best_match and min_diff < 0.001:
                    matched_pairs.append((col, best_match[1]))
                    print(f"✓ 能量值匹配: {col} -> {best_match[0]} ({best_match[1]:.8f} eV)")
                    matched = True
                else:
                    print(f"✗ 能量值不匹配: {col} (最接近的差值: {min_diff:.6f} eV)")
                    
            except ValueError:
                print(f"✗ 无法解析为能量值: {col}")
        
        if not matched:
            print(f"✗ 未匹配: {col}")
    
    print(f"\n成功匹配 {len(matched_pairs)} 个列")
    return matched_pairs

def get_colorscale_option():
    """选择颜色映射选项"""
    custom_colorscale = [
        [0.0, 'rgb(0, 0, 139)'],      # 深蓝色
        [0.15, 'rgb(0, 0, 255)'],     # 蓝色
        [0.3, 'rgb(0, 255, 255)'],    # 青色
        [0.45, 'rgb(0, 255, 0)'],     # 绿色
        [0.6, 'rgb(255, 255, 0)'],    # 黄色
        [0.75, 'rgb(255, 165, 0)'],   # 橙色
        [0.9, 'rgb(255, 0, 0)'],      # 红色
        [1.0, 'rgb(139, 69, 19)']     # 棕色
    ]
    return custom_colorscale

def create_rixs_with_custom_x_range(csv_file_path="./Andor/merged_file.csv",
                                    excitation_csv_path="./Andor/Excitation.csv",
                                    x_range=None,
                                    y_range=None,
                                    use_log_scale=True):
    """
    创建可自定义X轴范围的RIXS热力图
    
    参数:
    - csv_file_path: merged_file.csv文件路径
    - excitation_csv_path: Excitation.csv文件路径
    - x_range: X轴范围，例如 [530.0, 532.5]
    - y_range: Y轴范围，例如 [517.0, 531.0]
    - use_log_scale: 是否使用对数缩放
    """
    
    # 设置默认范围
    if x_range is None:
        x_range = [530.0, 532.5]
    if y_range is None:
        y_range = [517.0, 531.0]
    
    print(f"设置X轴范围: {x_range[0]:.2f} - {x_range[1]:.2f} eV")
    print(f"设置Y轴范围: {y_range[0]:.2f} - {y_range[1]:.2f} eV")
    
    # 读取CSV文件
    try:
        df = pd.read_csv(csv_file_path)
        print(f"成功读取 {csv_file_path}")
        print(f"数据形状: {df.shape}")
        print(f"列名: {list(df.columns)}")
    except Exception as e:
        print(f"读取merged_file.csv失败: {str(e)}")
        return None
    
    # 提取发射能量 (Y轴)
    emission_energy = df['X'].values
    
    # 从Excitation.csv获取激发能量映射
    energy_mapping = get_excitation_energy_mapping(excitation_csv_path)
    
    if not energy_mapping:
        print("无法获取激发能量映射，程序终止")
        return None
    
    # 提取数据列（除了X列）
    data_columns = [col for col in df.columns if col != 'X']
    
    # 智能匹配列名与激发能量
    matched_pairs = match_columns_with_energies(data_columns, energy_mapping)
    
    if not matched_pairs:
        print("错误: 没有找到匹配的激发能量数据")
        return None
    
    # 创建激发能量数组和强度数据
    excitation_energies = []
    intensity_data = []
    
    for col_name, energy in matched_pairs:
        excitation_energies.append(energy)
        intensity_data.append(df[col_name].values)
    
    excitation_energies = np.array(excitation_energies)
    intensity_data = np.array(intensity_data).T
    
    print(f"\n=== 数据信息 ===")
    print(f"激发能量范围: {excitation_energies.min():.2f} - {excitation_energies.max():.2f} eV")
    print(f"发射能量范围: {emission_energy.min():.2f} - {emission_energy.max():.2f} eV")
    print(f"强度数据形状: {intensity_data.shape}")
    
    # 限制Y轴范围
    y_mask = (emission_energy >= y_range[0]) & (emission_energy <= y_range[1])
    emission_filtered = emission_energy[y_mask]
    intensity_filtered = intensity_data[y_mask, :]
    
    print(f"过滤后数据形状: {intensity_filtered.shape}")
    
    # 数据预处理
    if use_log_scale:
        intensity_processed = np.log10(intensity_filtered + 1)
        colorbar_title = 'Log₁₀(Intensity + 1)'
    else:
        intensity_processed = intensity_filtered
        colorbar_title = 'Intensity'
    
    # 创建网格进行插值
    X_orig, Y_orig = np.meshgrid(excitation_energies, emission_filtered)
    
    # 创建插值网格
    x_interp_range = np.linspace(x_range[0], x_range[1], 150)
    y_interp_range = np.linspace(emission_filtered.min(), emission_filtered.max(), 200)
    X_interp, Y_interp = np.meshgrid(x_interp_range, y_interp_range)
    
    # 执行插值
    try:
        print("正在执行插值...")
        intensity_interp = griddata(
            (X_orig.ravel(), Y_orig.ravel()), 
            intensity_processed.ravel(), 
            (X_interp, Y_interp), 
            method='linear'
        )
        intensity_interp = np.nan_to_num(intensity_interp, nan=0)
        print("插值完成")
    except Exception as e:
        print(f"插值失败: {str(e)}")
        x_interp_range = excitation_energies
        y_interp_range = emission_filtered
        intensity_interp = intensity_processed
    
    # 获取颜色映射
    colorscale = get_colorscale_option()
    
    # 创建热力图
    fig = go.Figure(data=go.Heatmap(
        z=intensity_interp,
        x=x_interp_range,
        y=y_interp_range,
        colorscale=colorscale,
        colorbar=dict(
            title=colorbar_title,
            titleside='right',
            outlinecolor='black',
            outlinewidth=1
        ),
        showscale=True,
        hovertemplate='Excitation: %{x:.2f} eV<br>Emission: %{y:.2f} eV<br>Intensity: %{z:.2f}<extra></extra>'
    ))
    
    # 更新布局
    fig.update_layout(
        title={
            'text': 'SJD4_18031 - RIXS Map',
            'x': 0.5,
            'xanchor': 'center',
            'font': {'size': 20, 'color': 'black'}
        },
        xaxis=dict(
            title='Excitation energy / eV',
            titlefont=dict(size=18, color='black'),
            tickfont=dict(size=14, color='black'),
            showgrid=True,
            gridwidth=1,
            gridcolor='gray',
            showline=True,
            linewidth=2,
            linecolor='black',
            mirror=True,
            ticks='outside',
            tickwidth=2,
            ticklen=5,
            range=[x_range[0], x_range[1]]
        ),
        yaxis=dict(
            title='Emission energy / eV',
            titlefont=dict(size=18, color='black'),
            tickfont=dict(size=14, color='black'),
            showgrid=True,
            gridwidth=1,
            gridcolor='gray',
            showline=True,
            linewidth=2,
            linecolor='black',
            mirror=True,
            ticks='outside',
            tickwidth=2,
            ticklen=5,
            range=[y_range[0], y_range[1]]
        ),
        plot_bgcolor='white',
        height=600,
        width=800,
        margin=dict(l=80, r=80, t=80, b=80)
    )
    
    fig.show()
    
    print(f"\n=== 最终设置 ===")
    print(f"X轴显示范围: {x_range[0]:.2f} - {x_range[1]:.2f} eV")
    print(f"Y轴显示范围: {y_range[0]:.2f} - {y_range[1]:.2f} eV")
    print(f"强度缩放: {'对数' if use_log_scale else '线性'}")
    print(f"成功匹配了 {len(matched_pairs)} 个激发能量点")
    
    return fig

# 使用示例
if __name__ == "__main__":
    
    print("=== 使用Excitation.csv动态生成激发能量映射 ===")
    print("=== 智能匹配列名与激发能量 ===")
    
    # 创建默认图 - 使用线性缩放
    print("\n1. 创建默认范围的图 (530.0-532.5 eV) - 线性缩放")
    fig1 = create_rixs_with_custom_x_range(
        csv_file_path="./Andor/merged_file.csv",
        excitation_csv_path="./Andor/Excitation.csv",
        use_log_scale=False  # 使用线性缩放
    )
    
    if fig1 is not None:
        print("✓ 图表创建成功!")
    else:
        print("✗ 图表创建失败")
    
    print("\n=== 使用方法 ===")
    print("现在程序会智能匹配列名:")
    print("1. 如果列名是文件名格式，直接匹配")
    print("2. 如果列名是能量值，通过数值匹配")
    print("3. 自动处理不同的命名格式")
```

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from scipy.interpolate import griddata
import os

def get_excitation_energy_mapping(excitation_csv_path="./Andor/Excitation.csv"):
    """从Excitation.csv文件获取文件名到激发能量的映射关系"""
    try:
        # 读取Excitation.csv文件
        df = pd.read_csv(excitation_csv_path)
        
        # 检查必要的列是否存在
        if 'Filename' not in df.columns or 'BL 8 Energy' not in df.columns:
            print("Error: 'Filename' or 'BL 8 Energy' column not found in Excitation.csv")
            return {}
        
        # 创建映射字典
        energy_mapping = {}
        for index, row in df.iterrows():
            filename = row['Filename']
            energy = row['BL 8 Energy']
            
            # 处理文件名，添加-1D后缀（如果需要）
            if not filename.endswith('-1D'):
                filename_key = filename + '-1D'
            else:
                filename_key = filename
            
            energy_mapping[filename_key] = energy
        
        print(f"成功从 {excitation_csv_path} 读取了 {len(energy_mapping)} 个激发能量映射")
        print("映射关系:")
        for key, value in energy_mapping.items():
            print(f"  {key}: {value:.8f} eV")
        
        return energy_mapping
        
    except FileNotFoundError:
        print(f"错误: 找不到文件 {excitation_csv_path}")
        print("使用默认的硬编码映射关系...")
        return get_default_energy_mapping()
    except Exception as e:
        print(f"读取Excitation.csv时出错: {str(e)}")
        print("使用默认的硬编码映射关系...")
        return get_default_energy_mapping()

def get_default_energy_mapping():
    """默认的激发能量映射关系（作为备选）"""
    energy_mapping = {
        'SJD4_18031-00001-1D': 529.90157898,
        'SJD4_18031-00002-1D': 530.10417319,
        'SJD4_18031-00003-1D': 530.31199310,
        'SJD4_18031-00004-1D': 530.50475223,
        'SJD4_18031-00005-1D': 530.70780803,
        'SJD4_18031-00006-1D': 530.92626647,
        'SJD4_18031-00007-1D': 531.10930027,
        'SJD4_18031-00008-1D': 531.30263958,
        'SJD4_18031-00009-1D': 531.50630679,
        'SJD4_18031-00010-1D': 531.71013021,
        'SJD4_18031-00011-1D': 531.90900862,
        'SJD4_18031-00012-1D': 532.12845734,
        'SJD4_18031-00013-1D': 532.30210317,
        'SJD4_18031-00014-1D': 532.50653755,
        'SJD4_18031-00015-1D': 532.71112902,
        'SJD4_18031-00016-1D': 532.91587776
    }
    return energy_mapping

def match_columns_with_energies(data_columns, energy_mapping):
    """
    匹配merged_file.csv中的列名与激发能量映射
    处理列名可能是文件名或能量值的情况
    """
    matched_pairs = []
    
    # 创建能量值到文件名的反向映射
    energy_to_filename = {v: k for k, v in energy_mapping.items()}
    
    print(f"\n=== 列名匹配过程 ===")
    print(f"merged_file.csv中的列名: {data_columns}")
    
    for col in data_columns:
        matched = False
        
        # 方法1: 直接匹配文件名
        if col in energy_mapping:
            matched_pairs.append((col, energy_mapping[col]))
            print(f"✓ 直接匹配: {col} -> {energy_mapping[col]:.8f} eV")
            matched = True
        
        # 方法2: 尝试作为能量值匹配
        if not matched:
            try:
                energy_value = float(col)
                # 在能量映射中查找最接近的能量值
                best_match = None
                min_diff = float('inf')
                
                for filename, energy in energy_mapping.items():
                    diff = abs(energy - energy_value)
                    if diff < min_diff:
                        min_diff = diff
                        best_match = (filename, energy)
                
                # 如果差值小于0.001 eV，认为是匹配的
                if best_match and min_diff < 0.001:
                    matched_pairs.append((col, best_match[1]))
                    print(f"✓ 能量值匹配: {col} -> {best_match[0]} ({best_match[1]:.8f} eV)")
                    matched = True
                else:
                    print(f"✗ 能量值不匹配: {col} (最接近的差值: {min_diff:.6f} eV)")
                    
            except ValueError:
                print(f"✗ 无法解析为能量值: {col}")
        
        if not matched:
            print(f"✗ 未匹配: {col}")
    
    print(f"\n成功匹配 {len(matched_pairs)} 个列")
    return matched_pairs

def get_colorscale_option():
    """选择颜色映射选项"""
    custom_colorscale = [
        [0.0, 'rgb(0, 0, 139)'],      # 深蓝色
        [0.15, 'rgb(0, 0, 255)'],     # 蓝色
        [0.3, 'rgb(0, 255, 255)'],    # 青色
        [0.45, 'rgb(0, 255, 0)'],     # 绿色
        [0.6, 'rgb(255, 255, 0)'],    # 黄色
        [0.75, 'rgb(255, 165, 0)'],   # 橙色
        [0.9, 'rgb(255, 0, 0)'],      # 红色
        [1.0, 'rgb(139, 69, 19)']     # 棕色
    ]
    return custom_colorscale

def create_rixs_with_custom_x_range(csv_file_path="./Andor/merged_file.csv",
                                    excitation_csv_path="./Andor/Excitation.csv",
                                    x_range=None,
                                    y_range=None,
                                    use_log_scale=True):
    """
    创建可自定义X轴范围的RIXS热力图
    
    参数:
    - csv_file_path: merged_file.csv文件路径
    - excitation_csv_path: Excitation.csv文件路径
    - x_range: X轴范围，例如 [530.0, 532.5]
    - y_range: Y轴范围，例如 [517.0, 531.0]
    - use_log_scale: 是否使用对数缩放
    """
    
    # 先读取数据来确定默认范围
    try:
        df_temp = pd.read_csv(csv_file_path)
        energy_mapping_temp = get_excitation_energy_mapping(excitation_csv_path)
        
        if energy_mapping_temp:
            actual_x_min = min(energy_mapping_temp.values())
            actual_x_max = max(energy_mapping_temp.values())
            actual_y_min = df_temp['X'].min()
            actual_y_max = df_temp['X'].max()
            
            # 设置默认范围，如果没有指定的话
            if x_range is None:
                # 自动设置X范围，稍微扩展一点边界
                x_padding = (actual_x_max - actual_x_min) * 0.02  # 2% padding
                x_range = [actual_x_min - x_padding, actual_x_max + x_padding]
                print(f"自动检测X轴范围: {x_range[0]:.2f} - {x_range[1]:.2f} eV")
            
            if y_range is None:
                # 自动设置Y范围
                y_padding = (actual_y_max - actual_y_min) * 0.02  # 2% padding  
                y_range = [actual_y_min - y_padding, actual_y_max + y_padding]
                print(f"自动检测Y轴范围: {y_range[0]:.2f} - {y_range[1]:.2f} eV")
        else:
            # 如果无法自动检测，使用默认值
            if x_range is None:
                x_range = [530.0, 532.5]
            if y_range is None:
                y_range = [517.0, 531.0]
    except:
        # 如果出错，使用默认值
        if x_range is None:
            x_range = [530.0, 532.5]
        if y_range is None:
            y_range = [517.0, 531.0]
    
    print(f"设置X轴范围: {x_range[0]:.2f} - {x_range[1]:.2f} eV")
    print(f"设置Y轴范围: {y_range[0]:.2f} - {y_range[1]:.2f} eV")
    
    # 读取CSV文件
    try:
        df = pd.read_csv(csv_file_path)
        print(f"成功读取 {csv_file_path}")
        print(f"数据形状: {df.shape}")
        print(f"列名: {list(df.columns)}")
    except Exception as e:
        print(f"读取merged_file.csv失败: {str(e)}")
        return None
    
    # 提取发射能量 (Y轴)
    emission_energy = df['X'].values
    
    # 从Excitation.csv获取激发能量映射
    energy_mapping = get_excitation_energy_mapping(excitation_csv_path)
    
    if not energy_mapping:
        print("无法获取激发能量映射，程序终止")
        return None
    
    # 提取数据列（除了X列）
    data_columns = [col for col in df.columns if col != 'X']
    
    # 智能匹配列名与激发能量
    matched_pairs = match_columns_with_energies(data_columns, energy_mapping)
    
    if not matched_pairs:
        print("错误: 没有找到匹配的激发能量数据")
        return None
    
    # 创建激发能量数组和强度数据
    excitation_energies = []
    intensity_data = []
    
    for col_name, energy in matched_pairs:
        excitation_energies.append(energy)
        intensity_data.append(df[col_name].values)
    
    excitation_energies = np.array(excitation_energies)
    intensity_data = np.array(intensity_data).T
    
    print(f"\n=== 数据信息 ===")
    print(f"激发能量范围: {excitation_energies.min():.2f} - {excitation_energies.max():.2f} eV")
    print(f"发射能量范围: {emission_energy.min():.2f} - {emission_energy.max():.2f} eV")
    print(f"强度数据形状: {intensity_data.shape}")
    
    # 限制Y轴范围
    y_mask = (emission_energy >= y_range[0]) & (emission_energy <= y_range[1])
    emission_filtered = emission_energy[y_mask]
    intensity_filtered = intensity_data[y_mask, :]
    
    print(f"过滤后数据形状: {intensity_filtered.shape}")
    
    # 数据预处理
    if use_log_scale:
        intensity_processed = np.log10(intensity_filtered + 1)
        colorbar_title = 'Log₁₀(Intensity + 1)'
    else:
        intensity_processed = intensity_filtered
        colorbar_title = 'Intensity'
    
    # 创建网格进行插值
    X_orig, Y_orig = np.meshgrid(excitation_energies, emission_filtered)
    
    # 创建插值网格
    x_interp_range = np.linspace(x_range[0], x_range[1], 150)
    y_interp_range = np.linspace(emission_filtered.min(), emission_filtered.max(), 200)
    X_interp, Y_interp = np.meshgrid(x_interp_range, y_interp_range)
    
    # 执行插值
    try:
        print("正在执行插值...")
        intensity_interp = griddata(
            (X_orig.ravel(), Y_orig.ravel()), 
            intensity_processed.ravel(), 
            (X_interp, Y_interp), 
            method='linear'
        )
        intensity_interp = np.nan_to_num(intensity_interp, nan=0)
        print("插值完成")
    except Exception as e:
        print(f"插值失败: {str(e)}")
        x_interp_range = excitation_energies
        y_interp_range = emission_filtered
        intensity_interp = intensity_processed
    
    # 获取颜色映射
    colorscale = get_colorscale_option()
    
    # 创建热力图
    fig = go.Figure(data=go.Heatmap(
        z=intensity_interp,
        x=x_interp_range,
        y=y_interp_range,
        colorscale=colorscale,
        colorbar=dict(
            title=colorbar_title,
            titleside='right',
            outlinecolor='black',
            outlinewidth=1
        ),
        showscale=True,
        hovertemplate='Excitation: %{x:.2f} eV<br>Emission: %{y:.2f} eV<br>Intensity: %{z:.2f}<extra></extra>'
    ))
    
    # 更新布局
    fig.update_layout(
        title={
            'text': 'SJD4_18031 - RIXS Map',
            'x': 0.5,
            'xanchor': 'center',
            'font': {'size': 20, 'color': 'black'}
        },
        xaxis=dict(
            title='Excitation energy / eV',
            titlefont=dict(size=18, color='black'),
            tickfont=dict(size=14, color='black'),
            showgrid=True,
            gridwidth=1,
            gridcolor='gray',
            showline=True,
            linewidth=2,
            linecolor='black',
            mirror=True,
            ticks='outside',
            tickwidth=2,
            ticklen=5,
            range=[x_range[0], x_range[1]]
        ),
        yaxis=dict(
            title='Emission energy / eV',
            titlefont=dict(size=18, color='black'),
            tickfont=dict(size=14, color='black'),
            showgrid=True,
            gridwidth=1,
            gridcolor='gray',
            showline=True,
            linewidth=2,
            linecolor='black',
            mirror=True,
            ticks='outside',
            tickwidth=2,
            ticklen=5,
            range=[y_range[0], y_range[1]]
        ),
        plot_bgcolor='white',
        height=600,
        width=800,
        margin=dict(l=80, r=80, t=80, b=80)
    )
    
    # 尝试显示图表，如果失败则保存为HTML文件
    try:
        fig.show()
        print("✓ 图表已在浏览器中显示")
    except Exception as e:
        print(f"显示图表时出错: {e}")
        # 保存为HTML文件
        html_file = csv_file_path.replace('.csv', '_rixs_map.html')
        fig.write_html(html_file)
        print(f"✓ 图表已保存为HTML文件: {html_file}")
        
        # 保存为PNG文件（如果安装了kaleido）
        try:
            png_file = csv_file_path.replace('.csv', '_rixs_map.png')
            fig.write_image(png_file, width=800, height=600, scale=2)
            print(f"✓ 图表已保存为PNG文件: {png_file}")
        except Exception as img_error:
            print(f"保存PNG失败: {img_error}")
            print("提示: 安装 'pip install kaleido' 来支持PNG导出")
    
    print(f"\n=== 最终设置 ===")
    print(f"X轴显示范围: {x_range[0]:.2f} - {x_range[1]:.2f} eV")
    print(f"Y轴显示范围: {y_range[0]:.2f} - {y_range[1]:.2f} eV")
    print(f"强度缩放: {'对数' if use_log_scale else '线性'}")
    print(f"成功匹配了 {len(matched_pairs)} 个激发能量点")
    
    return fig

# 使用示例
if __name__ == "__main__":
    
    print("=== 使用Excitation.csv动态生成激发能量映射 ===")
    print("=== 智能匹配列名与激发能量 ===")
    
    # 创建默认图 - 使用线性缩放，自动检测范围
    print("\n1. 创建默认范围的图（自动检测数据范围）- 线性缩放")
    fig1 = create_rixs_with_custom_x_range(
        csv_file_path="./Andor/merged_file.csv",
        excitation_csv_path="./Andor/Excitation.csv",
        use_log_scale=False  # 使用线性缩放
    )
    
    # 如果你想要手动指定完整的数据范围，可以这样做：
    print("\n2. 手动指定完整数据范围的图:")
    print("   取消下面的注释来运行:")
    print("   fig2 = create_rixs_with_custom_x_range(")
    print("       csv_file_path='./Andor/merged_file.csv',")
    print("       excitation_csv_path='./Andor/Excitation.csv',")
    print("       x_range=[529.85, 532.95],  # 完整激发能量范围")
    print("       y_range=[517.0, 531.0],   # 完整发射能量范围")
    print("       use_log_scale=False")
    print("   )")
    
    # 取消注释下面的代码来运行完整范围的图
    # fig2 = create_rixs_with_custom_x_range(
    #     csv_file_path="./Andor/merged_file.csv",
    #     excitation_csv_path="./Andor/Excitation.csv",
    #     x_range=[529.85, 532.95],  # 完整激发能量范围
    #     y_range=[517.0, 531.0],   # 完整发射能量范围
    #     use_log_scale=False
    # )
    
    if fig1 is not None:
        print("✓ 图表创建成功!")
    else:
        print("✗ 图表创建失败")
    
    print("\n=== 使用方法 ===")
    print("现在程序会智能匹配列名:")
    print("1. 如果列名是文件名格式，直接匹配")
    print("2. 如果列名是能量值，通过数值匹配")
    print("3. 自动处理不同的命名格式")
```
